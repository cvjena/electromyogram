# Example Notebook - Scheme surface electrode locations

```python
import electromyogram as emg
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(1, 3, figsize=(30, 10))

ax[0].imshow(emg.plot_locations(scheme=emg.Fridlund(), shape=( 512,  512), draw_outerhull=True))
ax[1].imshow(emg.plot_locations(scheme=emg.Fridlund(), shape=(1024, 1024), draw_outerhull=True))
ax[2].imshow(emg.plot_locations(scheme=emg.Fridlund(), shape=(2048, 2048), draw_outerhull=True))
```

```python
fig, ax = plt.subplots(1, 3, figsize=(30, 10))

ax[0].imshow(emg.plot_locations(scheme=emg.Fridlund(), shape=( 512,  512), do_postprocess=False))
ax[1].imshow(emg.plot_locations(scheme=emg.Fridlund(), shape=(1024, 1024), do_postprocess=False))
ax[2].imshow(emg.plot_locations(scheme=emg.Fridlund(), shape=(2048, 2048), do_postprocess=False))
```

```python
fig, ax = plt.subplots(1, 3, figsize=(30, 10))

ax[0].imshow(emg.plot_locations(scheme=emg.Kuramoto(), shape=( 512,  512)))
ax[1].imshow(emg.plot_locations(scheme=emg.Kuramoto(), shape=(1024, 1024)))
ax[2].imshow(emg.plot_locations(scheme=emg.Kuramoto(), shape=(2048, 2048)))
```

# Overview - Fridlund vs Kuramoto

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
ax.axis("off")
ax.imshow(emg.plot_locations(scheme=emg.Fridlund(), shape=(1024, 1024), draw_outerhull=True))

plt.savefig("../files/locations_fridlund.jpg", dpi=100, bbox_inches="tight")
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
ax.axis("off")
ax.imshow(emg.plot_locations(scheme=emg.Kuramoto(), shape=(1024, 1024), draw_outerhull=True))

plt.savefig("../files/locations_kuramoto.jpg", dpi=100, bbox_inches="tight")
```
